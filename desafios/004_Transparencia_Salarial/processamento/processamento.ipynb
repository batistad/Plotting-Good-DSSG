```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv("PWIT Salary Transparency Project - Salaries Information.csv")
```

```python
df['salario_processado'] = pd.to_numeric(df['Monthly Gross Salary'].str.strip('€').str.replace(',| ', '', regex=True), errors='coerce')
df.loc[25, 'salario_processado'] = 2800 
df.loc[31, 'salario_processado'] = 4623.23
df.loc[131, 'salario_processado'] = 10229.62
df.loc[261, 'salario_processado'] = 4746.30
df.loc[269, 'salario_processado'] = 6430.08
df.loc[283, 'salario_processado'] = 13050.81
df.loc[426, 'salario_processado'] = 1690
df.loc[204, 'salario_processado'] = 1168.83
df.loc[248, 'salario_processado'] = 2126.83
df.loc[412, 'salario_processado'] = 1000
df.loc[236, 'salario_processado'] = 2000
```

```python
def set_bin_salario(row):
    if row["salario_processado"] < 2000:
        return "menos de 2000"
    elif row["salario_processado"] >= 2000 and  row["salario_processado"] < 4000:
        return "2000-3999"
    elif row["salario_processado"] >= 4000 and  row["salario_processado"] <= 6000:
        return "4000-6000"
    else:
        return "mais de 6000"

df = df.assign(bins_salario=df.apply(set_bin_salario, axis=1))
df.rename(columns = {'Age':'idade','Gender':'genero', 'Location':'local', 'Job Tittle' : 'titulo', 'Years of post-college professional work experience':"anos_de_experiencia", "Size of Company":"tamanho_da_empresa", "Monthly Gross Salary":"salario_bruto_mensal", "Other perks":"outros_beneficios" }, inplace = True)
df.replace({'genero': "Female"}, "mulher", inplace=True)
df.replace({'genero': "Male"}, "homem", inplace=True)
df["anos_de_experiencia"] = df["anos_de_experiencia"].str.replace("years of experience", "anos de experiência")
df["anos_de_experiencia"] = df["anos_de_experiencia"].str.replace("No experience", "Sem experiência")
df["tamanho_da_empresa"] = df["tamanho_da_empresa"].str.replace("people", "pessoas")

df
```

```python
df.to_csv("pwit_transparencia_salarial.csv")
```
